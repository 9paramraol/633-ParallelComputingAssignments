```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
def val_list(filename):
    file1 = open(filename, 'r')
    Lines = file1.readlines()
    ans = []
    for line in Lines:
        ans.append(float(line))
    return ans
```

```python
def new_lists(full_list):
    ans = {}
    n = len(full_list)
    for i in range(n):
        method = i%3
        size = (i//3)%7
        if (method not in ans):
            ans[method] = {}
        if (size not in ans[method]):
            ans[method][size] = []
        ans[method][size].append(full_list[i])
#     print(ans[0])
    return ans

def dict_to_list(a_dict):
    m1 = []; m2 = []; m3 = []
    for i in range(7):
        m1.append(a_dict[0][i])
        m2.append(a_dict[1][i])
        m3.append(a_dict[2][i])
    return m1,m2,m3
```

```python
nlist = val_list("data_16")
temp = new_lists(nlist)
a1,a2,a3 = dict_to_list(temp)
fig = plt.figure(figsize =(10, 7)) 
import statistics  
# Creating plot

plt.boxplot(a1,positions=[i for i in range(4,11)]) 
plt.plot([i for i in range(4,11)],[statistics.mean(i) for i in a1])
plt.boxplot(a2,positions=[i for i in range(4,11)])
plt.plot([i for i in range(4,11)],[statistics.mean(i) for i in a2])
plt.boxplot(a3,positions=[i for i in range(4,11)])
plt.plot([i for i in range(4,11)],[statistics.mean(i) for i in a3])
plt.xlabel('log(grid size)/2')
plt.ylabel('time in second')
# show plot 
plt.show() 
```
